# Data Preprocessing

## Treat Outlier

```python
def treat_outliers(data, feature):
    IQ1 = data[feature].quantile(0.25)
    IQ3 = data[feature].quantile(0.75)
    IQR = IQ3 - IQ1
    
    lower_bound = IQ1 - 1.5 * IQR
    upper_bound = IQ3 + 1.5 * IQR
    
    outliers = data[(data[feature] < lower_bound) | (data[feature] > upper_bound)]
    
    # Replace outliers with median value
    median = data[feature].median()
    data.loc[(data[feature] < lower_bound), feature] = median
    data.loc[(data[feature] > upper_bound), feature] = median
    
    return data
```

```python
train_dataset = treat_outliers(train_dataset, 'MaxOfUpperTRange')
test_dataset = treat_outliers(test_dataset, 'MaxOfUpperTRange')
```
